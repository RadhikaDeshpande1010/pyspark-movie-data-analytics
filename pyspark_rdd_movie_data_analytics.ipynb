```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local").appName("Colab Pyspark").getOrCreate()
print(spark)
```

```python
sc = spark.sparkContext
print(sc)
```

```python
file_name = sc.textFile("/content/sample_data/movie_data.txt")

# Extract header
header = file_name.first()

# Print header
print("Headers:", header)

# Remove header from RDD
rdd1 = file_name.filter(lambda row: row != header)

rdd1.collect()
```

```python
"""
1.  Extract only movie names using map().
"""
movie_name = rdd1.map(lambda movie_record: movie_record.split("|")) \
                 .map(lambda movie_fields: movie_fields[1])

movie_name.collect()
```

```python
"""
2.  Convert all movie names to uppercase.
"""
movieNameUpperCase = rdd1.map(lambda movie_record: movie_record.split("|")) \
                         .map(lambda movie_fields: movie_fields[1].upper())

movieNameUpperCase.collect()
```

```python
"""
3.  Convert all movie names to lowercase.
"""
movieNameLowerCase = rdd1.map(lambda movie_record: movie_record.split("|")) \
                         .map(lambda movie_fields: movie_fields[1].lower())

movieNameLowerCase.collect()
```

```python
"""
4.  Extract only release year and convert it to integer.
"""
movieReleaseYear = rdd1.map(lambda movie_record: movie_record.split("|")) \
                       .map(lambda movie_fields: int(movie_fields[3]))

movieReleaseYear.collect()
```

```python
"""
5.  Create key-value pair like (MovieName, Rating).
"""
movieNameAndRating = rdd1.map(lambda movie_record: movie_record.split("|")) \
                         .map(lambda movie_fields:
                              (movie_fields[1], float(movie_fields[5]))
                             )

movieNameAndRating.collect()
```

```python
"""
6.  Append “_MOVIE” to every movie name.
"""
movieNameAppend = rdd1.map(lambda movie_record: movie_record.split("|")) \
                      .map(lambda movie_fields: movie_fields[1] + "_MOVIE")

movieNameAppend.collect()
```

```python
"""
7.  Create key-value pair like (Genre, 1).
"""
genre_val = 1
movieGenre1 = rdd1.map(lambda movie_record: movie_record.split("|")) \
                  .map(lambda movie_fields: (movie_fields[2], genre_val))

movieGenre1.collect()
```

```python
"""
8.  Multiply each movie duration by 2.
"""
duration_value = 2
movieDurationBy2 = rdd1.map(lambda movie_record: movie_record.split("|")) \
                       .map(lambda movie_fields: int(movie_fields[4]) * duration_value)

movieDurationBy2.collect()
```

```python
"""
9.  Create key-value pair like (Country, Rating).
"""
movieCountryAndRating = rdd1.map(lambda movie_record: movie_record.split("|")) \
                            .map(lambda movie_fields:
                                 (movie_fields[6], float(movie_fields[5]))
                                )

movieCountryAndRating.collect()
```

```python
"""
10. Add 1.0 to each movie rating.
"""
rating_value = 1.0

movieRating = rdd1.map(lambda movie_record: movie_record.split("|")) \
                  .map(lambda movie_fields: float(movie_fields[5]) + rating_value)

movieRating.collect()
```

```python
"""
11. Filter movies where rating > 8.5.
"""
movieRatings = rdd1.map(lambda movie_record: movie_record.split("|")) \
                   .map(lambda movie_fields: (movie_fields[1], float(movie_fields[5]))) \
                   .filter(lambda movie_pair: movie_pair[1] > 8.5)

movieRatings.collect()
```

```python
"""
12. Filter movies from “Action” genre.
"""
actionGenreMovie = rdd1.map(lambda movie_record: movie_record.split("|")) \
                       .map(lambda movie_fields: (movie_fields[1], movie_fields[2])) \
                       .filter(lambda movie_pair: movie_pair[1] == "Action")

actionGenreMovie.collect()
```

```python
"""
13. Filter movies released after 2015.
"""
movieReleaseYear2015 = rdd1.map(lambda movie_record: movie_record.split("|")) \
                           .map(lambda movie_fields: (movie_fields[1], int(movie_fields[3]))) \
                           .filter(lambda movie_pair: movie_pair[1] > 2015)

movieReleaseYear2015.collect()
```

```python
"""
14. Filter movies where duration > 150 minutes.
"""
movieDuration150 = rdd1.map(lambda record: record.split("|")) \
                      .map(lambda fields: (fields[1], int(fields[4]))) \
                      .filter(lambda movie_duration: movie_duration[1] > 150)

movieDuration150.collect()
```

```python
"""
15. Filter movies from “India”.
"""
IndianMovies = rdd1.map(lambda movie_record: movie_record.split("|")) \
                   .map(lambda movie_fields: (movie_fields[1], movie_fields[6])) \
                   .filter(lambda movie_pair: movie_pair[1] == "India")

IndianMovies.collect()
```

```python
"""
16. Filter movie names starting with letter ‘T’.
"""
movieNameStartsWithT = rdd1.map(lambda record: record.split("|")) \
                          .map(lambda fields: fields[1]) \
                          .filter(lambda name: name.startswith("T"))

movieNameStartsWithT.collect()
```

```python
"""
17. Filter movies where rating between 8.0 and 8.5.
"""
movieRating = rdd1.map(lambda record: record.split("|")) \
                  .map(lambda fields: (fields[1], float(fields[5]))) \
                  .filter(lambda movie_rating: 8.0 <= movie_rating[1] <= 8.5)

movieRating.collect()
```

```python
"""
18. Filter movies from “Animation” genre.
"""
animationGenreMovie = rdd1.map(lambda record: record.split("|")) \
                      .map(lambda fields: (fields[1], fields[2])) \
                      .filter(lambda movie_genre: movie_genre[1] == "Animation")

animationGenreMovie.collect()
```

```python
"""
19. Filter movies where duration < 100 minutes.
"""
movieDuration100 = rdd1.map(lambda movie_record: movie_record.split("|")) \
                       .map(lambda movie_fields: (movie_fields[1], int(movie_fields[4]))) \
                       .filter(lambda movie_pair: movie_pair[1] < 100)

movieDuration100.collect()
```

```python
"""
20. Filter movies where name length > 10 characters.
"""
movieLength = rdd1.map(lambda record: record.split("|")) \
                  .map(lambda fields: fields[1]) \
                  .filter(lambda movie_length: len(movie_length) > 10)

movieLength.collect()
```
